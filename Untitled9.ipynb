```python
import sqlite3

conn = sqlite3.connect("sales.db")  # create SQLite DB
df.to_sql("sales", conn, if_exists="replace", index=False)
```

```python
query = """
SELECT "Product line", SUM(Total) AS Total_Sales
FROM sales
GROUP BY "Product line"
ORDER BY Total_Sales DESC
"""
df_sales = pd.read_sql(query, conn)
df_sales
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.barplot(data=df_sales, x='Total_Sales', y='Product line', palette='Set2')
plt.title('Total Sales by Product Line')
plt.xlabel('Total Sales')
plt.ylabel('Product Line')
plt.show()
```

```python
query2 = """
SELECT Payment, ROUND(AVG("gross income"), 2) AS Avg_Income
FROM sales
GROUP BY Payment
"""
df_income = pd.read_sql(query2, conn)
df_income
```

```python
plt.figure(figsize=(8,6))
sns.barplot(data=df_income, x='Payment', y='Avg_Income', palette='Set2')
plt.title('Average Gross Income by Payment Method', fontsize=16)
plt.xlabel('Payment Method', fontsize=14)
plt.ylabel('Average Gross Income', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
query3 = """
SELECT City, SUM(Total) as Total_Sales
FROM sales
GROUP BY City
ORDER BY Total_Sales DESC
"""
df_city = pd.read_sql(query3, conn)

plt.figure(figsize=(8,6))
sns.barplot(data=df_city, x='City', y='Total_Sales', palette='muted')
plt.title('Total Sales by City')
plt.show()
```

```python
query4 = """
SELECT "Customer type", ROUND(AVG(Rating), 2) as Avg_Rating
FROM sales
GROUP BY "Customer type"
"""
df_rating = pd.read_sql(query4, conn)

plt.figure(figsize=(6,5))
sns.barplot(data=df_rating, x='Customer type', y='Avg_Rating', palette='coolwarm')
plt.title('Average Rating by Customer Type')
plt.show()
```

```python
query6 = """
SELECT Gender, SUM("gross income") as Total_Income
FROM sales
GROUP BY Gender
"""
df_gender = pd.read_sql(query6, conn)

plt.figure(figsize=(6,5))
sns.barplot(data=df_gender, x='Gender', y='Total_Income', palette='pastel')
plt.title('Total Gross Income by Gender')
plt.show()
```

```python
query7 = """
SELECT Payment, COUNT(*) as Count
FROM sales
GROUP BY Payment
ORDER BY Count DESC
"""
df_payment = pd.read_sql(query7, conn)

plt.figure(figsize=(6,5))
sns.barplot(data=df_payment, x='Payment', y='Count', palette='deep')
plt.title('Most Used Payment Methods')
plt.show()
```
